```python
import seaborn as sns
df = sns.load_dataset("tips")
```

```python
df.head()
```

Two numerical data (Use scatter plot)**bold text**

```python
sns.scatterplot(x='tip', y='total_bill', hue=df['sex'],data=df,style='smoker',size='size')
```

one numerical and one categorical data (Use bar plot)**bold text**

```python
sns.barplot(x='day', y='tip', data=df,hue='sex')
```

```python
sns.boxplot(x='tip',y='sex', data=df,hue='smoker')
```

```python
sns.distplot(df['tip'],hist=False)
sns.distplot(df['total_bill'],hist=False)
```

Categerica-Categerical (HeatMap)

```python
sns.heatmap(pd.crosstab(df['sex'],df['day']),annot=True)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.lineplot(x='tip',y='total_bill',data=df)
```
